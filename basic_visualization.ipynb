# Data Visialization

## This is a new file of code showing a basic visualization

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [3, 7, 8, 5, 10]

# Create a basic line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Data Line")

# Add labels and title
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Matplotlib Visualization")
plt.legend()

# Show the plot
plt.show()
```
